```python
import tarfile
dName = 'resample/'
dir_ = 'resample/12_18_resample_cM/'
tar = tarfile.open('resample/12_18_resample_cM.tar.gz', "r:gz")
tar.extractall(dName)
tar.close()
```

```python
import glob
import numpy as np
d = {}
def readFile(fname):
    piD, pi, D, B, nSites = np.loadtxt(fname).T
    return piD, pi, D, B, nSites
# column order
# TODO * instead of YRI
def readFiles(m, initial):
    for fname in glob.glob(dir_+'*.%spi_d.thresh.*.cM.wsize_*.txt'%initial):
        pop, n = fname.split('/')[-1].split('.%spi_d.thresh.'%initial)
        thresh, n = n.split('.cM.wsize_')
        wSize = float(n.split('.txt')[0])
        thresh = float(thresh)
        k = (pop, m, thresh, wSize)
        d[k] = readFile(fname)
        print(fname)
        assert len(d[k]) == 5
readFiles('X','x')
readFiles('A','')
modes = ['X', 'A']
def ths(pop, c=2):
    return sorted(set([k[c] for k in d.keys() if k[0] == pop]))
```

```python
import numpy as np
pop = 'YRI'
print ths(pop)
# find nan's
for k in d.keys():
    #for l in d[k]:
    for i in xrange(5):
        if np.isnan(d[k][i]).any():
            print 'nan: ', k, i, np.count_nonzero(~np.isnan(d[k][i]))
            # TODO REMOVE
            new0 = [v for v in d[k][0] if not np.isnan(v)]
            assert i == 0
            d[k] = (new0,d[k][1],d[k][2],d[k][3],d[k][4])
print 'done, %d, %d, %d'%(len(ths(pop)), len(ths(pop,3)),len(d))
print ths(pop,3)
```

```python
import matplotlib as mpl
import seaborn
import random
plt = mpl.pyplot
# set high DPI on figures globally
mpl.rcParams['figure.dpi'] = 500
mpl.rcParams['axes.titlesize'] = 16
mpl.rcParams['axes.labelsize'] = 14
mpl.rcParams['xtick.labelsize'] = 14
mpl.rcParams['ytick.labelsize'] = 14
mpl.rcParams['legend.fontsize'] = 12

fSize = 1000000.0
fRange = [x for x in ths(pop) if x <=0.8]
def plot_cm(pop):
    # plot log10 neutral site counts per bin
    #TODO plt.figure(figsize=(10, 12))
    plt.figure(figsize=(10, 18))
    ax = plt.subplot(611)
    clrs = {'X':'darkturquoise','A':'mediumorchid'}
    style = {'X':'-','A':'-'}
    # number of sites
    for m in modes:
        wsize, c = fSize, clrs[m]
        mean = np.log10([np.average(d[pop, m, th, wsize][4]) for th in fRange])
        upper = np.log10([np.percentile(d[pop, m, th, wsize][4], 97.5) for th in fRange])
        lower = np.log10([np.percentile(d[pop, m, th, wsize][4], 2.5) for th in fRange])
        # plt.plot(ths(pop), lower, color=c,linestyle=style[m])
        # plt.plot(ths(pop), upper, color=c,linestyle=style[m])
        plt.fill_between(fRange, lower, upper, color=c, alpha='0.5')
        plt.plot(fRange, mean, color=c,linestyle=style[m])
    plt.ylabel('Number of neutral\nsites (log-10)')
    plt.xticks(color='white')
    plt.xlim([0,0.8])
    plt.text(0.02,5.25,'A',fontsize=16,color='black')
    # average McVicker's B
    ax = plt.subplot(612)
    plt.xlim([0,0.8])
    plt.xticks(color='white')
    labels = {'X':'X','A':'Autosomes'}
    for m in modes:
        wsize, c = fSize, clrs[m]
        mean  = [np.average(d[pop, m, th, wsize][3])/1000.0 for th in fRange]
        upper = [np.percentile(d[pop, m, th, wsize][3], 97.5)/1000.0 for th in fRange]
        lower = [np.percentile(d[pop, m, th, wsize][3], 2.5)/1000.0 for th in fRange]
        plt.plot(fRange, mean, color=c,label=labels[m],linestyle=style[m])
        #plt.plot(fRange, upper, color=c, label=labels[m],linestyle=style[m])
        plt.fill_between(fRange, lower, upper, color=c, alpha='0.5')
        plt.legend(loc=10)
    plt.ylabel('Average McVicker\'s B')
    plt.text(0.02,0.625,'B',fontsize=16,color='black')
    # average pi/D
    ax = plt.subplot(613)
    plt.xlim([0,0.8])
    plt.xticks(color='white')
    for m in modes:
        wsize, c = fSize, clrs[m]
        mean  = [np.average(d[pop, m, th, wsize][0]) for th in fRange]
        upper = [np.percentile(d[pop, m, th, wsize][0], 97.5) for th in fRange]
        # print 'pi/D at .2 * %s'%m,  np.average(d[pop, m, 0.2, wsize][0])
        # print 'INVERSE pi/D at .2 * %s'%m,  1.0/np.average([1.0/x for x in d[pop, m, 0.2, wsize][0]])
        # print '*.2 %s :'%m, np.percentile(d[pop, m, 0.2, wsize][0], 97.5),np.percentile(d[pop, m, 0.2, wsize][0], 2.5),np.percentile(d[pop, m, 0.2, wsize][0], 97.5) - np.percentile(d[pop, m, 0.2, wsize][0], 2.5), np.std(d[pop, m, 0.2, wsize][0])
        # print '*.3 %s :'%m, np.percentile(d[pop, m, 0.3, wsize][0], 97.5) , np.percentile(d[pop, m, 0.3, wsize][0], 2.5),np.percentile(d[pop, m, 0.3, wsize][0], 97.5) - np.percentile(d[pop, m, 0.3, wsize][0], 2.5), np.std(d[pop, m, 0.3, wsize][0])
        lower = [np.percentile(d[pop, m, th, wsize][0], 2.5) for th in fRange]
        #plt.plot(ths(pop), lower, color=c,linestyle=style[m])
        #plt.plot(ths(pop), upper, color=c,linestyle=style[m])
        plt.plot(fRange, mean, color=c,linestyle=style[m])
        plt.fill_between(fRange, lower, upper, color=c, alpha='0.5')
    plt.ylabel(r'$\pi/D$')
    plt.text(0.02,0.0095,'C',fontsize=16,color='black')

    # X:A ratio
    ax = plt.subplot(614)
    plt.xlim([0,0.8])
    listsX = [d[pop, 'X', th, fSize][0] for th in fRange]
    listsA = [d[pop, 'A', th, fSize][0] for th in fRange]
    lists = [[random.choice(listsX[i])/random.choice(listsA[i]) for _ in xrange(10000)] for i in xrange(len(fRange))]
    mean = [np.average(l) for l in lists]
    upper = [np.percentile(l, 97.5) for l in lists]
    lower = [np.percentile(l, 2.5) for l in lists]
    print 'X:A at .2Cm: ', [(mean[i],lower[i],upper[i]) for i in range(len(mean)) if ths(pop)[i] == 0.2]
    plt.plot(fRange, mean, color='g')
    plt.fill_between(fRange, lower, upper, color='green', alpha='0.5')
    plt.ylabel('X:A Ratio')
    plt.xlabel('Lower threshold for neutral sites (cM)')
    plt.text(0.02,0.575,'D',fontsize=16,color='black')
    return
    # average pi
    ax = plt.subplot(615)
    plt.xlim([0,1])
    plt.xticks(color='white')
    labels = {'X':'X','A':'Autosomes'}
    for m in modes:
        mean  = [np.average(d[pop, m, th][1]) for th in ths(pop)]
        upper = [np.percentile(d[pop, m, th][1], 95) for th in ths(pop)]
        lower = [np.percentile(d[pop, m, th][1], 5.0) for th in ths(pop)]
        plt.plot(ths(pop), mean, color=colors[m], label=labels[m])
        plt.fill_between(ths(pop), lower, upper, color=colors[m], alpha='0.5')
        plt.legend(loc=10)
    plt.ylabel('Average pi')
    # average D
    ax = plt.subplot(616)
    plt.xlim([0,1])
    plt.xticks(color='white')
    labels = {'X':'X','A':'Autosomes'}
    for m in modes:
        mean  = [np.average(d[pop, m, th][2]) for th in ths(pop)]
        upper = [np.percentile(d[pop, m, th][2], 97.5) for th in ths(pop)]
        lower = [np.percentile(d[pop, m, th][2], 2.5) for th in ths(pop)]
        plt.plot(ths(pop), mean, color='k', label=labels[m])
        plt.fill_between(ths(pop), lower, upper, color=colors[m], alpha='0.5')
        plt.legend(loc=10)
    plt.ylabel('Average D')
    plt.show()
    plt.close()

plot_cm('YRI')

```
